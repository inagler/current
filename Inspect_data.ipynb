```python
import numpy as np          # fundamental package for scientific computing
import glob                 # return all file paths that match a specific pattern
import matplotlib.pyplot as plt 
import os
```

```python
#files = glob.glob('~/phase1_CONDA/timeseries/demeaned/*.npy')
files = glob.glob(os.path.expanduser('~/phase1_CONDA/timeseries/demeaned/*.npy'))
time = np.arange(1850,1850+3012/12,1/12)
```

```python
variable = 5
data = np.load(files[variable])
```

```python
# Generate example data
num_series = data.shape[1]

# Set the initial shift value
shift = 0

# Create a large vertical figure
fig, ax = plt.subplots(figsize=(8, 20), dpi=300)

# Loop through each time series, plot with the desired shift, and add row numbers
for i in range(num_series):
    ax.plot(time, data[:,i] + shift, label=f"Row {i + 1}", linewidth=0.5)
    ax.text(time[-1]- 40, data[i, -1]+shift+20, f"member {i + 1}", verticalalignment='center', fontsize=8)
    shift += 40  # Increase the shift for the next series

# Set labels and title if needed
ax.set_xlabel('Time')
ax.set_title('Monthly Time Series ' + files[variable][49:][:-20])
ax.yaxis.set_ticks([])  # Remove y-axis ticks

# Add vertical grid lines at the specified tick locations
ax.grid(True, which='both', linestyle='-', linewidth=0.5, color='gray')

# Show the plot
plt.show()
```

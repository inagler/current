```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
path = '/Data/gfi/share/ModData/CESM2_LENS2/ocean/monthly/comp/'
temp_file = 'temp_anomaly.nc'
salt_file = 'salt_anomaly.nc'
shf_file = 'shf_anomaly.nc'
```

```python
ds_temp = xr.open_dataset(path+temp_file).roll(nlon=+100)
ds_temp.TEMP.sel(z_t=slice(5.0000000e+02, 1.050000e+04)).mean('z_t').plot()
```

```python
ds_salt = xr.open_dataset(path+salt_file).roll(nlon=+100)
ds_salt.SALT.sel(z_t=slice(5.0000000e+02, 1.050000e+04)).mean('z_t').plot()
```

```python
ds_shf = xr.open_dataset(path+shf_file).roll(nlon=+100)
ds_shf.SHF.plot()
```
